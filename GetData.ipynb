## 1. Thu thập các đường dẫn thuộc mỗi chuyên mục từ trang web https://vnexpress.net

#### - Viết hàm lấy tất cả đường dẫn của bài báo trong trang web

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

# Đường dẫn đến trang web gốc
URL = 'https://vnexpress.net'


#from urllib.error import HTTPError, URLError


def _get_links(url):
        links = []
        html = urlopen(url).read()
        soup = BeautifulSoup(html, 'html.parser')

        # Regex tìm tất cả các URL trong chuỗi HTML
        urls = set(re.findall(
            r'(?:http|https|ftp)://[\w\-_]+(?:\.[\w\-_]+)+(?:[\w\-\.,@?^=%&:/~\+#]*[\w\-\@?^=%&/~\+#])?',
            str(soup)
        ))

        for url in urls:
            # Lọc bài báo hợp lệ
            if url != URL and url.startswith(URL) and not url.startswith(URL + '/tac-gia') and url.endswith('.html'):
                links.append(url)
        return links
```

```python
from urllib.error import HTTPError, URLError
# Danh sách các chuyên mục
categories = [ 'thoi-su', 'the-gioi', 'kinh-doanh']

# Danh sách các đường dẫn đến các bài báo theo từng chuyên mục
article_url = {}

for category in categories:
    article_url[category] = []

    # lấy 10 trang đầu tiên của mỗi chuyên mục
    for i in range(1, 11):
        url = '{}/{}-p{}'.format(URL, category, str(i))
        article_url[category] += _get_links(url)

    # Loại bỏ các đường dẫn trùng lặp
    article_url[category] = list(set(article_url[category]))

    # Hiển thị số lượng bài báo trong mỗi chuyên mục
    print('Số lượng bài báo trong chuyên mục {}: {}'.format(category, len(article_url[category])))

# Lấy riêng các đường dẫn đến trong chuyên mục 'goc-nhin' theo từng mục con
goc_nhin = ['y-te-suc-khoe','chinh-tri-chinh-sach', 'van-hoa-loi-song','moi-truong']
article_url['goc-nhin'] = []

for sub_cate in goc_nhin:
    url = '{}/{}/{}-p{}'.format(URL, 'goc-nhin', sub_cate, i)
    article_url['goc-nhin'] += _get_links(url)

article_url['goc-nhin'] = list(set(article_url['goc-nhin']))
print('goc-nhin', len(article_url['goc-nhin']))
```

```python
def _get_content(url):
        content = ''
        html = urlopen(url).read()
        soup = BeautifulSoup(html, 'html.parser')

        div_content = soup.select_one('.page-detail .container')
        if div_content:
            # Mô tả
            description = div_content.find_all('p', {'class': 'description'})
            if description:
                description = description[-1]
                text_description = description.get_text()
                location = description.find('span', {'class': 'location-stamp'})
                if location:
                    content = text_description[len(location.get_text()):]
                else:
                    content = text_description

            # Nội dung bài báo
            detail = div_content.find('article', {'class': 'fck-detail'})
            if detail is not None:
                p_normal = detail.find_all('p', {'class': 'Normal'})
                if len(p_normal) > 0:
                    for p in p_normal:
                        p_text = p.get_text()
                        if p_text != '' and not p_text.startswith('>>') and p_text[-1] in ['.', '!', '?']:
                            content += ' ' + p.get_text()

        return re.sub(r'\n', '', content)
```

```python
import os

sentence = []
uppercase = "AÁÀẢÃẠĂẮẰẲẴẶÂẤẦẨẪẬBCCDEÉÈẺẼẸÊẾỀỂỄỆFGHIÍÌỈĨỊKLMNOÓÒỎÕỌÔỐỒỔỖỘƠỚỜỞỠỢPQRSTUÚÙỦŨỤƯỨỪỬỮỰVXY"

for category in article_url.keys():
    current_path = 'data/categories/' + category + '/'
    os.makedirs(current_path, exist_ok=True)

    count = 0
    for i in range(min(len(article_url[category]), 100)):
        content = _get_content(article_url[category][i])
        if content != '':
            count += 1
            with open(f'{current_path}/{category}_{count}.txt', 'w', encoding='utf-8') as f:
                lines = re.sub(r'(?<=[.?!])(?=[\'"“”‘’]*[' + uppercase + '-])', ' ', content)
                lines = re.split(r'(?<=[.?!])(?=[\])' + uppercase + '0-9-])', content)

                for line in lines:
                    f.write(line + '\n')
                    sentence.append(line)
    print(category, 'đã có', count, 'bài báo')

sentence = list(set(sentence))
print('Tổng số câu đã làm sạch:', len(sentence))
```

```python
# file ngữ liệu
file_corpus = open('data/corpus.txt', 'w', encoding='utf-8')

count = 0

for sent in sentence:
    # Xóa bộ ký tự thừa
    sent = re.sub(r'\u200b', '', sent)

    file_corpus.write(sent.strip() + '\n')

    count += 1

file_corpus.close()

print('Tổng số câu trong file ngữ liệu:', count)
```
